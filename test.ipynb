```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
X = np.random.rand(100, 2)  # 100 samples with 2 features
y = np.random.randint(0, 2, 100)  # Binary target variable (0 or 1)

# Define the labels for the positive and negative classes
pos_label = "Class 1"
neg_label = "Class 0"

# Create a scatter plot of the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", marker="o")

# Set the labels of the x and y axes
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Create a legend with the labels for the positive and negative classes
plt.legend(handles=[plt.plot([], [], marker="o", color="red", label=pos_label)[0],
                    plt.plot([], [], marker="o", color="blue", label=neg_label)[0]])

# Show the plot
plt.show()
```
